```python
"""2) What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot"""
```

```python
#scatter plot is used for plotting the coordinates in the graph which are the joining points of values of x and y.
```

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,marker="<",color="black")
plt.xlabel("This is x")
plt.ylabel("This is y")
plt.title("This is title")
```

```python
"""3)  Why is the subplot() function used? Draw four line plots using the subplot() function"""
```

```python
#With the subplot() function you can draw multiple plots in one figure
```

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line 1",size="7")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Line 2",size="7")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Line 3",size="7")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Line 4",size="7")

plt.show()
```

```python
"""4)  What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot."""
```

```python
#The purpose of a bar graph is to convey relational information quickly in a visual manner.
```

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.show
```

```python
plt.barh(company,profit)
plt.show
```
